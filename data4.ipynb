```python
df = pd.read_csv('vaccine_covid_1.csv', parse_dates=['date'])
# , index_col='location'
df.columns
```

```python
to_drop = []
for c in df.columns:
    if 'smoothed' in c:
        to_drop.append(c)    
    elif 'per'in c:
        to_drop.append(c)
    elif 'excess'in c:
        to_drop.append(c)
    elif 'index'in c:
        to_drop.append(c)
    elif 'unit'in c:
        to_drop.append(c)
    elif 'weekly'in c:
        to_drop.append(c)         
    elif 'rate'in c:
        to_drop.append(c)         
    elif 'total'in c:
        to_drop.append(c)
    elif 'test'in c:
        to_drop.append(c)
    elif 'vaccinat'in c:
        to_drop.append(c)

to_drop.remove('people_vaccinated')                  
df = df.drop(to_drop, axis=1)

to_drop1 = []
for i in range(10,20):
    to_drop1.append(df.columns[i])
df = df.drop(to_drop1, axis=1)
df.columns   
```
